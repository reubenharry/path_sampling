## 2D double well example

```python
from path_sampling import plot_path, sample_sde


def make_double_well_potential_2D(v):
    def potential(x):
        print(x.shape, "x shape")
        # return None
        """
        Double well potential
        """
    
        assert x.shape == (2,)
        return jnp.sum(v*(x**2 - 1)**2, axis=-1)
    return potential

make_double_well_potential_2D(v=1.0)(jnp.array([0.5, 0.5]))

def uref_2d(x,t):
    return -jax.grad(make_double_well_potential_2D(v=5.0))(x)

path, time = sample_sde(
    b=uref_2d,
    W=lambda x, key: jax.random.normal(key, shape=(2,)),
    rho=lambda key: jnp.ones((2,)),
    dt=0.01,
    num_steps=100,
    key=jax.random.PRNGKey(0)
)

# def plot_path(path, potential):

#     # plot path in 2D
#     plt.figure()
#     plt.plot(path[:, 0], path[:, 1])
#     plt.title("Sampled Path")
#     # plot heatmap of the potential make_double_well_potential
#     v = 1.0
#     x = np.linspace(-2, 2, 100)
#     y = np.linspace(-2, 2, 100)
#     X, Y = np.meshgrid(x, y)
#     print(X.shape)
#     # use vmap to apply the potential to the grid
#     Z = jax.vmap(lambda x,y: potential(jnp.array([x,y])))(X.reshape(-1), Y.reshape(-1)).reshape(X.shape)
#     plt.contourf(X, Y, Z, levels=50)

plot_path(path=path, time=time, potential=make_double_well_potential_2D(1.0), i=0, label=f"s: 0")
```

```python
hyperparams={'dt': 0.01, 'num_steps': 100, 'num_training_steps' : 1000}

time = np.arange(0,hyperparams['num_steps'])*hyperparams['dt']


force = jax.grad(lambda x: -jnp.sum(x**4 - 8 * x**2))


def numerical_laplacian(arr, dt):

    arr = jnp.concatenate([jnp.array([arr[0]]), arr, jnp.array([arr[-1]])])
    fst_diff = (arr[1:] - arr[:-1])
    snd_diff = (fst_diff[1:] - fst_diff[:-1])
    jax.debug.print("snd_diff {x}", x=snd_diff)
    return snd_diff

def spde_integrator(x, force, sigma=1, num_steps=10):
    ds = 0.01


    def step(x):

        # second_difference = (first_difference[:-1] - ) / hyperparams['dt']
        laplace_t_x = numerical_laplacian(x, dt=hyperparams['dt'])
        term2 = -jax.vmap(force)(x)*jax.vmap(jax.grad(force))(x)
        term3 = -(sigma**2)*jax.vmap(jax.grad(jax.grad(force)))(x)
        noise = jnp.sqrt(2)*jax.random.normal(key, shape=x.shape)

        return x + ds*(1/(sigma**2))*(laplace_t_x + term2 + term3) + jnp.sqrt(ds)*noise
    
    for i in range(num_steps):
        x = step(x)
        # print(x.shape)

    return x

# spde_integrator(
#     x=jnp.array([1.,2.,1.,4.,5.,1.,1.]),
#     force=force,
#     ).shape



initial = jnp.linspace(-1,1, 100)
# initial
```
