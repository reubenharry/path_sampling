```python
import numpy as np
import matplotlib.pyplot as plt

# Define the potential function
def double_well(x, y):
    # x : [100, 100]
    # y : [100, 100]
    # we want to do, for z=[x,y], (x**2 - 1)**2 
    return (x**2 - 1)**2 + (y**2 - 1)**2
    # return (x**2 - 1)**2 + 4 * (y**2 - 1)**2

# Generate x and y coordinates
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Calculate the potential values
Z = double_well(X, Y)

# Plot the potential
plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Double Well Potential')
plt.show()
```
